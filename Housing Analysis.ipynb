```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_csv(r"C:\Users\Vikas\Downloads\train.csv")
test_df = pd.read_csv(r"C:\Users\Vikas\Downloads\test.csv")
```

```python
mean_sale_price = train_df['SalePrice'].mean()
```

```python
print("Mean Sale Price:", mean_sale_price)
```

```python
median_sale_price = train_df['SalePrice'].median()
print("Median Sale Price:", median_sale_price)
```

```python
std_sale_price = train_df['SalePrice'].std()
print("Std Sale Price:", std_sale_price)
```

```python
count_per_mszoning = train_df['MSZoning'].value_counts()
print("Count per MSzoning:", count_per_mszoning)
```

```python
mean_lot_area_by_mszoning = train_df.groupby('MSZoning')['LotArea'].mean()
print("Mean LotArea by MSzoning:", mean_lot_area_by_mszoning)
```

```python
median_lot_frontage_by_mszoning = train_df.groupby('MSZoning')['LotFrontage'].median()
print("Median LotFrontage by MSzoning:", median_lot_frontage_by_mszoning)
```

```python
sale_price_by_sale_type = train_df.groupby('SaleType')['SalePrice'].describe()
print("Sale Price Distribution by Sale Type:",sale_price_by_sale_type)
```

```python
sale_price_by_sale_condition = train_df.groupby('SaleCondition')['SalePrice'].describe()
print("Sale Price Distribution by Sale Condition:", sale_price_by_sale_condition)
```

```python
count_by_year_sold = train_df['YrSold'].value_counts()
print("Count of Houses by Year Sold:", count_by_year_sold)
```

```python
avg_sale_price_by_year_sold = train_df.groupby('YrSold')['SalePrice'].mean()
print("Average Sale Price by Year Sold:", avg_sale_price_by_year_sold)
```

```python
plt.figure(figsize=(15, 10))
```

```python
# Mean and Median Sale Price
plt.subplot(2, 3, 1)
plt.bar(['Mean', 'Median'], [mean_sale_price, median_sale_price], color=['blue', 'green'])
plt.title('Mean and Median Sale Price')
plt.ylabel('Sale Price')
```

```python
# Standard Deviation of Sale Price
plt.subplot(2, 3, 2)
plt.bar(['Standard Deviation'], [std_sale_price], color='red')
plt.title('Standard Deviation of Sale Price')
plt.ylabel('Sale Price')
```

```python
# Count of Houses per MSZoning
plt.subplot(2, 3, 3)
count_per_mszoning.plot(kind='bar', color='purple')
plt.title('Count of Houses per MSZoning')
plt.ylabel('Count')
plt.xticks(rotation=45)
```

```python
# Mean Lot Area by MSZoning
plt.subplot(2, 3, 4)
mean_lot_area_by_mszoning.plot(kind='bar', color='orange')
plt.title('Mean Lot Area by MSZoning')
plt.ylabel('Lot Area')
plt.xticks(rotation=45)
```

```python
# Median Lot Frontage by MSZoning
plt.subplot(2, 3, 5)
median_lot_frontage_by_mszoning.plot(kind='bar', color='cyan')
plt.title('Median Lot Frontage by MSZoning')
plt.ylabel('Lot Frontage')
plt.xticks(rotation=45)
```

```python
# Count of Houses by Year Sold
plt.subplot(2, 3, 6)
count_by_year_sold.plot(kind='bar', color='magenta')
plt.title('Count of Houses by Year Sold')
plt.ylabel('Count')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Sale Price Distribution by Sale Type
plt.figure(figsize=(12, 6))
sns.boxplot(x='SaleType', y='SalePrice', data=train_df)
plt.title('Sale Price Distribution by Sale Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sale Price Distribution by Sale Condition
plt.figure(figsize=(12, 6))
sns.boxplot(x='SaleCondition', y='SalePrice', data=train_df)
plt.title('Sale Price Distribution by Sale Condition')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average Sale Price by Year Sold
plt.figure(figsize=(12, 6))
avg_sale_price_by_year_sold.plot(kind='line', marker='o', color='blue')
plt.title('Average Sale Price by Year Sold')
plt.ylabel('Sale Price')
plt.xlabel('Year Sold')
plt.grid(True)
plt.show()
```
